### 3.1 Data Wrangling for 'Zomato Restaurant names and Metadata' Dataset

```python
# 1. Drop duplicate rows

print(f"Shape before dropping duplicates: {df_metadata.shape}")

df_metadata.drop_duplicates(inplace=True)
print(f"Shape after dropping duplicates: {df_metadata.shape}")
```

```python
# 2. Check and handle missing values

print("\nMissing values count before handling:")
display(df_reviews.isnull().sum().sort_values(ascending=False))

# Drop rows with missing values for 'Review', 'Rating'
df_reviews.dropna(subset=['Review', 'Rating'], inplace=True)

# Fill 'Reviewer' missing values with 'unknown reviewer'
df_reviews['Reviewer'].fillna('unknown reviewer', inplace=True)

# Fill 'Time' missing values with its mode
mode_time = df_reviews['Time'].mode()[0]
df_reviews['Time'].fillna(mode_time, inplace=True)

# Fill 'Metadata' missing values with '0 reviews, 0 followers'
df_reviews['Metadata'].fillna('0 reviews, 0 followers', inplace=True)

print("\nMissing values count after handling:")
display(df_reviews.isnull().sum().sort_values(ascending=False))
```

```python
# 3. Standardize restaurant names, review, and Metadata columns (lowercase, remove extra spaces)

df_reviews['Restaurant'] = df_reviews['Restaurant'].astype(str).str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
df_reviews['Review'] = df_reviews['Review'].astype(str).str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
df_reviews['Metadata'] = df_reviews['Metadata'].astype(str).str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
```

```python
# 4. Remove leading/trailing whitespaces and whitespaces after commas from text columns

# For 'Restaurant' (already handled in step 2 for extra spaces, strip covers leading/trailing)
# For 'Reviewer'
df_reviews['Reviewer'] = df_reviews['Reviewer'].astype(str).str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
# For 'Review' (already handled in step 2 for extra spaces, strip covers leading/trailing)
```

```python
# 5. Convert Rating to numeric datatype

# Convert rating to string first, then remove 'like' if present (e.g., '3.5 like') and convert to numeric
df_reviews['Rating'] = df_reviews['Rating'].astype(str).str.lower().str.replace(' like', '', regex=False)
df_reviews['Rating'] = pd.to_numeric(df_reviews['Rating'], errors='coerce')

print("\nRating column info after conversion:")
display(df_reviews['Rating'].describe())
```

```python
# 6. Validate Rating values (between 1-5)

invalid_ratings = df_reviews[(df_reviews['Rating'] < 1) | (df_reviews['Rating'] > 5) | (df_reviews['Rating'].isnull())]   # rating should be betwwen 1 to 5
print(f"\nNumber of rows with invalid ratings (outside 1-5 or NaN): {len(invalid_ratings)}")

# Optionally, remove or correct invalid ratings. For this task, we will remove them.
if not invalid_ratings.empty:
    print("Removing rows with invalid ratings...")
    df_reviews.drop(invalid_ratings.index, inplace=True)
    print(f"Shape after removing invalid ratings: {df_reviews.shape}")

print("\nRating column info after validation:")
display(df_reviews['Rating'].describe())
```

```python
# 8. Create review_length feature

df_reviews['review_length'] = df_reviews['Review'].apply(lambda x: len(str(x).split()) if pd.notnull(x) else 0)

print("\nReview_length column description:")
display(df_reviews['review_length'].describe())
```

```python
import re

# 9. Extract reviewer_review_count and reviewer_follower_count from Metadata

def extract_counts(metadata_str):
    if pd.isna(metadata_str) or metadata_str == '0 reviews, 0 followers':
        return 0, 0

    metadata_str = str(metadata_str).lower()

    # Try to match 'x reviews' and 'y followers'
    # Make 's' optional in 'reviews' and 'followers' to handle singular forms
    reviews_match = re.search(r'(\d+)(?:k)? reviews?', metadata_str)
    followers_match = re.search(r'(\d+)(?:k)? followers?', metadata_str)

    reviews = 0
    if reviews_match:
        reviews_str = reviews_match.group(1)
        if 'k' in reviews_match.group(0):
            reviews = float(reviews_str) * 1000
        else:
            reviews = float(reviews_str)

    followers = 0
    if followers_match:
        followers_str = followers_match.group(1)
        if 'k' in followers_match.group(0):
            followers = float(followers_str) * 1000
        else:
            followers = float(followers_str)

    return int(reviews), int(followers)


df_reviews[['reviewer_review_count', 'reviewer_follower_count']] = df_reviews['Metadata'].apply(lambda x: pd.Series(extract_counts(x)))

print("\nReviewer counts description:")
display(df_reviews[['reviewer_review_count', 'reviewer_follower_count']].describe())
```

### **4.13 Top 10 influencial food critics by critic score**

*** creating a new dataframe named critic_df for this chart according to the project's specific requirements.

```python
from sklearn.preprocessing import MinMaxScaler

# Create a reviewer-level DataFrame (critic_df)
# Aggregate relevant features for each reviewer
critic_df = df_reviews.groupby('Reviewer').agg(
    reviewer_follower_count=('reviewer_follower_count', 'max'), # Max to get the actual count
    reviewer_review_count=('reviewer_review_count', 'max')     # Max to get the actual count
).reset_index()

# Handle cases where a reviewer might have 0 followers or reviews (e.g., from '0 reviews, 0 followers' filling)
critic_df['reviewer_follower_count'] = critic_df['reviewer_follower_count'].fillna(0)
critic_df['reviewer_review_count'] = critic_df['reviewer_review_count'].fillna(0)

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Normalize 'reviewer_follower_count' and 'reviewer_review_count'
critic_df[['normalized_followers', 'normalized_reviews']] = scaler.fit_transform(
    critic_df[['reviewer_follower_count', 'reviewer_review_count']]
)

print("Critic DataFrame created and features normalized:")
display(critic_df.head())
display(critic_df[['reviewer_follower_count', 'reviewer_review_count', 'normalized_followers', 'normalized_reviews']].describe())
```

```python
# Calculate 'critic_score' with 60% weight for followers and 40% for review count
critic_df['critic_score'] = (0.6 * critic_df['normalized_followers']) + (0.4 * critic_df['normalized_reviews'])

# Identify the Top 10 critics based on critic score
top_10_critics = critic_df.sort_values(by='critic_score', ascending=False).head(10)

print("Top 10 Critics based on critic_score:")
display(top_10_critics[['Reviewer', 'reviewer_follower_count', 'reviewer_review_count', 'critic_score']])
```

### **5.1 Hypothesis Test 1 : Restaurant Cost vs. Average Rating**

**Hypothesis:** Is there a significant linear relationship between the cost of a restaurant and its average rating?

*   **Null Hypothesis (H0):** There is no significant linear correlation between restaurant cost and average rating (ρ = 0).
*   **Alternative Hypothesis (H1):** There is a significant linear correlation between restaurant cost and average rating (ρ ≠ 0).

```python
from scipy.stats import pearsonr

# Extract the 'Cost' and 'Average_Rating' columns
cost = df_merged_cost_rating['Cost']
ratings = df_merged_cost_rating['Average_Rating']

# Perform Pearson correlation test
correlation_coefficient, p_value = pearsonr(cost, ratings)

print(f"Pearson Correlation Coefficient: {correlation_coefficient:.4f}")
print(f"P-value: {p_value:.10f}")

# Interpret the p-value
alpha = 0.05
print(f"\nSignificance Level (alpha): {alpha}")

if p_value < alpha:
    print("Since the p-value is less than the significance level, we reject the null hypothesis (H0).")
    print("Conclusion: There is a statistically significant linear relationship between restaurant cost and average rating.")
else:
    print("Since the p-value is greater than the significance level, we fail to reject the null hypothesis (H0).")
    print("Conclusion: There is no statistically significant linear relationship between restaurant cost and average rating.")
```

Conclusion: Since the p-value is less than 0.05, the null hypothesis is rejected. There is a statistically significant positive relationship between restaurant cost and average rating. The Pearson correlation coefficient (r = 0.4223) indicates a moderate positive correlation, suggesting that higher-cost restaurants tend to receive slightly higher ratings on average.

### a) **Which statistical test have I done to obtain P-Value?**

Pearson Correlation Test

### b) **Why did I choose the specific statistical test?**

Pearson Correlation Test was used because both Restaurant Cost and Average Rating are numerical variables. The test helps measure the strength and direction of their linear relationship and determines whether the observed correlation is statistically significant.

### **5.2 Hypothesis Test 2: Review Length vs. Rating Categories**

**Hypothesis:** Is there a significant difference in average review length across different rating categories (1.0, 1.5, 2.0, ..., 5.0)?

*   **Null Hypothesis (H0):** The mean review length is the same across all rating categories (μ1 = μ2 = ... = μk).
*   **Alternative Hypothesis (H1):** At least one rating category has a different mean review length.

```python
from scipy.stats import f_oneway

# Group review lengths by rating categories
# Create a list of review lengths for each unique rating
rating_groups = [df_reviews['review_length'][df_reviews['Rating'] == rating] for rating in df_reviews['Rating'].unique()]

# Perform one-way ANOVA (Analysis of Variance) test
f_statistic, p_value = f_oneway(*rating_groups)

print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:e}")

# Interpret the p-value
alpha = 0.05
print(f"\nSignificance Level (alpha): {alpha}")

if p_value < alpha:
    print("Since the p-value is less than the significance level, we reject the null hypothesis (H0).")
    print("Conclusion: There is a statistically significant difference in mean review length across at least two rating categories.")
else:
    print("Since the p-value is greater than the significance level, we fail to reject the null hypothesis (H0).")
    print("Conclusion: There is no statistically significant difference in mean review length across rating categories.")
```

ANOVA compares:

(Variation between rating groups / Variation within rating groups)

Formula:

F = (Between-group variance / Within-group variance)

so , If F is small -> The average review lengths of different rating groups are very similar.

and if F is large -> The differences between rating-group means are much larger than normal random variation.
in this case its the second one.

**conclusion:** Since the p-value (4.027113e-64) is less than 0.05, the null hypothesis is rejected. The ANOVA test (F = 40.191) indicates a statistically significant difference in average review length across rating categories. This suggests that customers with different ratings tend to write reviews of different lengths.

### **a) Which statistical test have you done?**

One-Way ANOVA Test

### **b) Why did you choose this test?**

One-Way ANOVA was used because review length is a numerical variable and ratings consist of multiple groups. The test determines whether the mean review length differs significantly across rating categories.

### **6.2 outlier detection**

```python
# for df_metadata

print("\n--- Outlier Detection using IQR for df_metadata ---")

outlier_results = []

# Automatically detect numerical columns
numerical_cols = df_metadata.select_dtypes(include=np.number).columns

for col in numerical_cols:
    Q1 = df_metadata[col].quantile(0.25)
    Q3 = df_metadata[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df_metadata[(df_metadata[col] < lower_bound) | (df_metadata[col] > upper_bound)]
    num_outliers = len(outliers)

    outlier_results.append({
        'Column': col,
        'Q1': f"{Q1:.2f}",
        'Q3': f"{Q3:.2f}",
        'IQR': f"{IQR:.2f}",
        'Lower Bound': f"{lower_bound:.2f}",
        'Upper Bound': f"{upper_bound:.2f}",
        'Number of Outliers': num_outliers
    })

# Display results in a clean tabular format
outlier_df = pd.DataFrame(outlier_results)
display(outlier_df.set_index('Column'))

print("\n--- End of Outlier Detection ---")
```

```python
import nltk
import re
import contractions
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download NLTK resources (only need to run once)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('punkt_tab', quiet=True) # Added explicit download for punkt_tab

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Define preprocessing functions

# 1. Expand contractions
def expand_contractions_text(text):
    return contractions.fix(text)

# 2. Remove punctuation
def remove_punctuation(text):
    return re.sub(r'[^a-zA-Z\s]', ' ', text) # Keep alphanumeric and whitespace (unicode range for A-Z, a-z, space)

# 3. Remove URLs
def remove_urls(text):
    return re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

# 4. Remove words containing digits
def remove_digits_words(text):
    return re.sub(r'\b\w*\d\w*\b', '', text)

# 5. Remove stopwords
def remove_stopwords_text(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(filtered_tokens)

# 6. Apply lemmatization
def lemmatize_text(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(lemmatized_tokens)

# 7. Remove extra whitespaces
def remove_extra_whitespace(text):
    return re.sub(r'\s+', ' ', text).strip()

# Create a new column 'clean_review' by copying the 'Review' column
# This ensures the original 'Review' column remains untouched
df_reviews['clean_review'] = df_reviews['Review'].copy()

print("Applying text preprocessing steps to 'df_reviews['Review']'...")

# Apply preprocessing steps sequentially

# 1. Expand contractions
df_reviews['clean_review'] = df_reviews['clean_review'].apply(expand_contractions_text)

# 2. Convert text to lowercase (ensuring consistent casing after contractions)
df_reviews['clean_review'] = df_reviews['clean_review'].str.lower()

# 3. Remove URLs
df_reviews['clean_review'] = df_reviews['clean_review'].apply(remove_urls)

# 4. Remove words containing digits
df_reviews['clean_review'] = df_reviews['clean_review'].apply(remove_digits_words)

# 5. Remove punctuation
df_reviews['clean_review'] = df_reviews['clean_review'].apply(remove_punctuation)

# 6. Remove stopwords
df_reviews['clean_review'] = df_reviews['clean_review'].apply(remove_stopwords_text)

# 7. Apply lemmatization
df_reviews['clean_review'] = df_reviews['clean_review'].apply(lemmatize_text)

# 8. Remove extra whitespaces
df_reviews['clean_review'] = df_reviews['clean_review'].apply(remove_extra_whitespace)

print("Text preprocessing completed. New column 'clean_review' created.")

# Display before-and-after examples for 5 reviews
print("\n--- Before and After Text Preprocessing Examples (5 Reviews) ---")

# Sample 5 random reviews for display
sample_reviews = df_reviews.sample(n=5, random_state=42) # Use random_state for reproducibility

# Create a temporary DataFrame for displaying the comparison
display_df = pd.DataFrame({
    'Review (Original from df_reviews["Review"])': sample_reviews['Review'],
    'Cleaned Review': sample_reviews['clean_review']
})

display(display_df)

print("\nShape of df_reviews after adding 'clean_review' column:", df_reviews.shape)
print("First 5 rows of df_reviews with 'Review' and 'clean_review':")
display(df_reviews[['Review', 'clean_review']].head())
display(df_reviews.head())
```

```python
# Import TfidfVectorizer for text vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
# - stop_words='english': Removes common English stopwords (e.g., 'the', 'is') that don't carry much meaning.
# - max_features=5000: Limits the vocabulary to the top 5000 most frequent terms, reducing dimensionality.
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)

print("Applying TF-IDF Vectorization to 'clean_review' column...")

# Fit and transform the 'clean_review' column to create the TF-IDF feature matrix
# This converts the text data into a sparse numerical matrix.
tfidf_matrix = tfidf_vectorizer.fit_transform(df_reviews['clean_review'])

print("TF-IDF Vectorization completed.")

# Display the shape of the generated TF-IDF matrix
# The shape represents (number of documents, number of unique terms/features).
print(f"\nShape of TF-IDF matrix: {tfidf_matrix.shape}")

# Get the feature names (terms) from the TF-IDF vectorizer
feature_names = tfidf_vectorizer.get_feature_names_out()

# Display the top 20 generated features (terms) to understand the vocabulary
print("\nTop 20 generated TF-IDF features (terms):")
# Sort features by their inverse document frequency (IDF) for insight into important terms
# Note: This is an approximation of importance, actual term importance is in the matrix values.
top_20_features = sorted(feature_names, key=lambda x: tfidf_vectorizer.idf_[tfidf_vectorizer.vocabulary_[x]], reverse=True)[:20]
for i, feature in enumerate(top_20_features):
    print(f"{i+1}. {feature}")

# Store the vectorized data (tfidf_matrix) for future sentiment analysis and clustering tasks
# It's good practice to save this directly if it's a large matrix or save the vectorizer.
# For now, we will keep it in memory as 'tfidf_matrix'.
print("\nTF-IDF matrix stored as 'tfidf_matrix' for further use.")
```

```python
# Analyze Feature Importance using Average TF-IDF Scores

print("\n--- Analyzing Top 20 Most Important TF-IDF Features ---")

# Get feature names from the TF-IDF vectorizer
feature_names = tfidf_vectorizer.get_feature_names_out()

# Calculate the average TF-IDF score for each feature across all documents.
# tfidf_matrix is a sparse matrix, so .mean(axis=0) will compute the mean for each column (feature).
# .A1 converts the sparse matrix result (which is effectively 1D) to a dense 1-D NumPy array for mean calculation.
import numpy as np
average_tfidf_scores = tfidf_matrix.mean(axis=0).A1

# Create a DataFrame to store feature names and their average TF-IDF scores
# This helps in easy sorting and display.
tfidf_feature_importance_df = pd.DataFrame({
    'feature': feature_names,
    'average_tfidf_score': average_tfidf_scores
})

# Sort the features by their average TF-IDF score in descending order
# This ranks features by their overall importance across the dataset.
tfidf_feature_importance_df = tfidf_feature_importance_df.sort_values(
    by='average_tfidf_score', ascending=False
).reset_index(drop=True)

# Display the top 20 most important TF-IDF terms
print("\nTop 20 Most Important TF-IDF Terms (by Average TF-IDF Score):")
display(tfidf_feature_importance_df.head(20))

# Confirm that the original tfidf_matrix and tfidf_vectorizer remain unchanged
print("\nTF-IDF matrix and vectorizer remain unchanged.")
```

TF-IDF (Term Frequency–Inverse Document Frequency) vectorization was used to convert textual reviews into numerical features. TF-IDF assigns higher importance to informative terms while reducing the influence of commonly occurring words. It was chosen because it is computationally efficient, interpretable, and highly effective for sentiment analysis and text classification tasks.

#### **6.6.3 Skewness Analysis**

Skewness measures the asymmetry of the probability distribution of a real-valued random variable about its mean. A positive skew indicates a longer tail to the right (right-skewed), and a negative skew indicates a longer tail to the left (left-skewed). Understanding skewness helps in deciding whether a feature needs transformation to achieve a more symmetric distribution, which can improve the performance of many statistical models.

```python
import numpy as np

# Calculate skewness for each selected feature
print("Skewness for Selected Features:")
skewness_results = df_metadata[transformation_features].skew()
display(skewness_results)

# Determine if transformation might be required based on skewness
# A common rule of thumb is that if skewness is outside the range [-0.5, 0.5] or [-1, 1], transformation might be beneficial.
skewness_threshold = 0.5 # Using a common threshold for moderate skewness

print(f"\nTransformation Requirement Analysis (Threshold: |skewness| > {skewness_threshold}):")
for feature in transformation_features:
    skew = skewness_results[feature]
    if abs(skew) > skewness_threshold:
        print(f"- Feature '{feature}' (Skewness: {skew:.4f}): Transformation might be required.")
    else:
        print(f"- Feature '{feature}' (Skewness: {skew:.4f}): Transformation likely not required.")

print("\nNote: No transformations are applied in this step, only the requirement is assessed.")
```

### **Which dimensionality reduction technique have you used and why?**

Principal Component Analysis (PCA) was used for dimensionality reduction. PCA transforms the original features into a smaller set of uncorrelated principal components while preserving most of the variance in the dataset. In this project, PCA reduced the feature space from 85 features to 34 principal components while retaining 95% of the total variance, making the data more efficient for clustering.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

print("\n--- Dimensionality Reduction for Clustering Features (PCA) ---")

# 1. Combine all scaled clustering features into a single DataFrame
# This includes 'Cost_log', 'cuisine_count', 'collection_count_log' from clustering_features_scaled,
# and the one-hot encoded cuisine and collection features.
# Ensure index alignment for concatenation.

# First, confirm the existence of necessary DataFrames for concatenation
if 'clustering_features_scaled' not in locals():
    print("Error: 'clustering_features_scaled' DataFrame not found. Please run previous scaling steps.")
elif 'df_cuisines_encoded' not in locals():
    print("Error: 'df_cuisines_encoded' DataFrame not found. Please run previous encoding steps.")
elif 'df_collections_encoded' not in locals():
    print("Error: 'df_collections_encoded' DataFrame not found. Please run previous encoding steps.")
else:
    # Concatenate the scaled numerical features with the encoded categorical features
    full_clustering_features_scaled = pd.concat(
        [clustering_features_scaled, df_cuisines_encoded, df_collections_encoded],
        axis=1
    )

    print("Shape of full clustering features before PCA:")
    print(full_clustering_features_scaled.shape)

    # 2. Apply PCA
    # Initialize PCA. We don't specify n_components yet to analyze all components.
    pca = PCA()

    # Fit PCA on the scaled data and transform it
    pca.fit(full_clustering_features_scaled)

    # 3. Display the cumulative explained variance ratio
    explained_variance_ratio_cumulative = np.cumsum(pca.explained_variance_ratio_)

    print("\nCumulative Explained Variance Ratio:")
    for i, cum_var in enumerate(explained_variance_ratio_cumulative):
        if i < 5 or (i > len(explained_variance_ratio_cumulative) - 5):
            print(f"  Component {i+1}: {cum_var:.4f}")
        elif i == 5:
            print("  ...")

    # 4. Determine the minimum number of principal components required to retain at least 95% variance
    n_components_95_var = np.where(explained_variance_ratio_cumulative >= 0.95)[0][0] + 1
    print(f"\nMinimum number of principal components to retain 95% variance: {n_components_95_var}")

    # 5. Create a PCA-transformed DataFrame using the selected number of components
    pca_final = PCA(n_components=n_components_95_var)
    clustering_features_pca = pca_final.fit_transform(full_clustering_features_scaled)

    # Convert the PCA-transformed array back to a DataFrame for easier handling
    pca_columns = [f'PC_{i+1}' for i in range(n_components_95_var)]
    df_clustering_pca = pd.DataFrame(clustering_features_pca, columns=pca_columns, index=df_metadata.index)

    # 6. Display the shape before and after PCA
    print(f"\nShape after PCA (retaining {n_components_95_var} components): {df_clustering_pca.shape}")
    print("First 5 rows of PCA-transformed DataFrame:")
    display(df_clustering_pca.head())

    # 7. Plot the cumulative explained variance curve
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(explained_variance_ratio_cumulative) + 1), explained_variance_ratio_cumulative, marker='o', linestyle='--')
    plt.axvline(x=n_components_95_var, color='r', linestyle=':', label=f'{n_components_95_var} Components for 95% Variance')
    plt.axhline(y=0.95, color='g', linestyle='--', label='95% Variance Threshold')
    plt.title('Cumulative Explained Variance by Principal Components')
    plt.xlabel('Number of Principal Components')
    plt.ylabel('Cumulative Explained Variance Ratio')
    plt.grid(True)
    plt.legend()
    plt.show()

    print("\n--- PCA Dimensionality Reduction Completed ---")
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print("--- Training Multinomial Naive Bayes Classifier ---")

# Initialize the Multinomial Naive Bayes classifier
# This model is well-suited for classification with discrete features (like word counts or TF-IDF).
mnb_classifier = MultinomialNB()

# Train the model on the training data (TF-IDF features and sentiment labels)
mnb_classifier.fit(X_train, y_train)

print("Multinomial Naive Bayes model trained successfully.")

# Predict sentiment labels on the test set
y_pred = mnb_classifier.predict(X_test)

print("--- Model Evaluation ---")

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted') # Use 'weighted' for imbalanced classes
recall = recall_score(y_test, y_pred, average='weighted')       # Use 'weighted' for imbalanced classes
f1 = f1_score(y_test, y_pred, average='weighted')               # Use 'weighted' for imbalanced classes

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision (Weighted): {precision:.4f}")
print(f"Recall (Weighted): {recall:.4f}")
print(f"F1 Score (Weighted): {f1:.4f}")

# Display the classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Display the confusion matrix using a heatmap for better visualization
print("\nConfusion Matrix:")
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=mnb_classifier.classes_, yticklabels=mnb_classifier.classes_)
plt.title('Confusion Matrix for Multinomial Naive Bayes')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Create a bar chart for Accuracy, Precision, Recall, and F1 Score
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
scores = [accuracy, precision, recall, f1]

plt.figure(figsize=(10, 6))
sns.barplot(x=metrics, y=scores, palette='viridis')
plt.title('Multinomial Naive Bayes Model Performance Metrics')
plt.ylabel('Score')
plt.ylim(0, 1) # Metrics are between 0 and 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print("--- Multinomial Naive Bayes Model Evaluation Completed ---")
```

```python
# Import necessary libraries for hyperparameter tuning and evaluation
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print("\n--- Hyperparameter Tuning for Logistic Regression Classifier ---")

# Initialize the Logistic Regression classifier with fixed parameters not being tuned.
# random_state ensures reproducibility.
# max_iter is kept sufficiently high for convergence.
log_reg_tuned_classifier = LogisticRegression(random_state=42, max_iter=1000)

# Define the parameter grid for GridSearchCV.
# 'C': Inverse of regularization strength; smaller values specify stronger regularization.
# 'solver': Algorithm to use in the optimization problem.
#   'liblinear' is good for small datasets and L1/L2 regularization.
#   'lbfgs' is a good default for multiclass problems and supports L2 regularization.
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs']
}

# Set up GridSearchCV.
# 'cv=5' specifies 5-fold cross-validation for robust evaluation of each parameter combination.
# 'scoring="f1_weighted"' is chosen as the primary evaluation metric, suitable for imbalanced datasets.
# 'n_jobs=-1' utilizes all available CPU cores for faster computation.
# 'verbose=1' provides a progress report during the search.
grid_search_log_reg = GridSearchCV(
    estimator=log_reg_tuned_classifier,
    param_grid=param_grid,
    cv=5,
    scoring='f1_weighted',
    n_jobs=-1,
    verbose=1 # Display progress
)

# Fit GridSearchCV on the training data.
# This process will explore all parameter combinations and perform cross-validation to find the best set.
grid_search_log_reg.fit(X_train, y_train)

print("Hyperparameter tuning completed for Logistic Regression.")

# Display the best parameters found by GridSearchCV.
print(f"\nBest Parameters (Logistic Regression): {grid_search_log_reg.best_params_}")

# Display the best cross-validation F1-score achieved with the optimal parameters.
print(f"Best Cross-Validation F1-Score (Logistic Regression): {grid_search_log_reg.best_score_:.4f}")

# Retrieve the best estimator (model) from the grid search results.
# This model is trained with the identified optimal hyperparameters.
best_log_reg_model = grid_search_log_reg.best_estimator_
print("\nBest Estimator (Optimized Logistic Regression Model):")
print(best_log_reg_model)

print("\n--- Evaluation of Tuned Logistic Regression Model ---")

# Predict sentiment labels on the test set using the best, optimized model.
y_pred_log_reg_tuned = best_log_reg_model.predict(X_test)

# Calculate evaluation metrics for the tuned Logistic Regression model.
accuracy_log_reg_tuned = accuracy_score(y_test, y_pred_log_reg_tuned)
precision_log_reg_tuned = precision_score(y_test, y_pred_log_reg_tuned, average='weighted')
recall_log_reg_tuned = recall_score(y_test, y_pred_log_reg_tuned, average='weighted')
f1_log_reg_tuned = f1_score(y_test, y_pred_log_reg_tuned, average='weighted')

print(f"Accuracy (Tuned Logistic Regression): {accuracy_log_reg_tuned:.4f}")
print(f"Precision (Weighted, Tuned Logistic Regression): {precision_log_reg_tuned:.4f}")
print(f"Recall (Weighted, Tuned Logistic Regression): {recall_log_reg_tuned:.4f}")
print(f"F1 Score (Weighted, Tuned Logistic Regression): {f1_log_reg_tuned:.4f}")

# Display the classification report for a detailed breakdown of metrics per class.
print("\nClassification Report (Tuned Logistic Regression Model):")
print(classification_report(y_test, y_pred_log_reg_tuned))

# Display the confusion matrix using a heatmap for a clear visualization of true vs. predicted labels.
conf_matrix_log_reg_tuned = confusion_matrix(y_test, y_pred_log_reg_tuned)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_log_reg_tuned, annot=True, fmt='d', cmap='Blues',
            xticklabels=best_log_reg_model.classes_, yticklabels=best_log_reg_model.classes_)
plt.title('Confusion Matrix for Tuned Logistic Regression')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

print("--- Tuned Logistic Regression Model Evaluation Completed ---")
```

### **7.4.1 Creating ResturantSentimentPipeline class**

```python
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

class RestaurantSentimentPipeline:
    """A pipeline for predicting restaurant sentiment, encapsulating a TF-IDF vectorizer and a Logistic Regression model."""

    def __init__(self, tfidf_vectorizer: TfidfVectorizer, sentiment_model: LogisticRegression):
        """
        Initializes the RestaurantSentimentPipeline.

        Args:
            tfidf_vectorizer (TfidfVectorizer): The trained TF-IDF vectorizer.
            sentiment_model (LogisticRegression): The trained Logistic Regression sentiment model.
        """
        self.tfidf_vectorizer = tfidf_vectorizer
        self.sentiment_model = sentiment_model

    def predict_sentiment(self, text: str) -> str:
        """
        Predicts the sentiment for a given raw review text.

        Args:
            text (str): The raw review string to predict sentiment for.

        Returns:
            str: The predicted sentiment label (e.g., 'Positive', 'Negative', 'Neutral').
        """
        # Transform the raw text using the stored TF-IDF vectorizer
        # The input text needs to be in a list format for the transform method
        text_tfidf = self.tfidf_vectorizer.transform([text])

        # Predict sentiment using the stored Logistic Regression model
        predicted_sentiment = self.sentiment_model.predict(text_tfidf)[0]

        return predicted_sentiment
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print("--- Determining Optimal Number of Clusters for PCA-Transformed Feature Set ---")

# --- Step 1: Elbow Method for PCA Features ---
# The Elbow Method looks for the 'elbow' in the plot of inertia versus the number of clusters.
# Inertia is the sum of squared distances of samples to their closest cluster center.
# A smaller inertia generally means more compact clusters. The 'elbow' point indicates
# where adding more clusters no longer significantly decreases inertia.
print("\n--- Step 1: Elbow Method ---")
inertia_pca = []
k_range_pca = range(2, 11) # Test K values from 2 to 10

for k in k_range_pca:
    # Initialize KMeans with the current number of clusters (k).
    # random_state ensures reproducibility of the results.
    # n_init='auto' (or a specific integer like 10) is used to explicitly set the number
    # of times the k-means algorithm will be run with different centroid seeds.
    kmeans_pca = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_pca.fit(df_clustering_pca) # Fit on the PCA-transformed data
    inertia_pca.append(kmeans_pca.inertia_)

# Display inertia values for each K to observe the trend numerically.
print("Inertia values for each K (PCA-Transformed Features):")
for k, val in zip(k_range_pca, inertia_pca):
    print(f"K={k}: {val:.2f}")

# Plot the Elbow Curve to visually identify the optimal K.
plt.figure(figsize=(10, 6))
sns.lineplot(x=list(k_range_pca), y=inertia_pca, marker='o')
plt.title('Elbow Method for Optimal K (PCA-Transformed Features)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(list(k_range_pca))
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# --- Step 2: Silhouette Analysis for PCA Features ---
# The Silhouette Score measures how similar an object is to its own cluster (cohesion)
# compared to other clusters (separation). The score ranges from -1 to 1.
# A high value indicates that the object is well matched to its own cluster and poorly
# matched to neighboring clusters. Scores near 0 indicate overlapping clusters.
print("\n--- Step 2: Silhouette Analysis ---")
silhouette_scores_pca = []

for k in k_range_pca:
    kmeans_pca = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels_pca = kmeans_pca.fit_predict(df_clustering_pca)
    # Calculate the silhouette score using the PCA-transformed data and the generated cluster labels.
    score_pca = silhouette_score(df_clustering_pca, cluster_labels_pca)
    silhouette_scores_pca.append(score_pca)

# Display silhouette scores for each K to observe which K yields the highest score.
print("Silhouette Scores for each K (PCA-Transformed Features):")
for k, score in zip(k_range_pca, silhouette_scores_pca):
    print(f"K={k}: {score:.4f}")

# Plot Silhouette Score vs Number of Clusters to visually identify the peak score.
plt.figure(figsize=(10, 6))
sns.lineplot(x=list(k_range_pca), y=silhouette_scores_pca, marker='o', color='green')
plt.title('Silhouette Score for Optimal K (PCA-Transformed Features)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.xticks(list(k_range_pca))
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

print("--- Optimal Number of Clusters Determination Completed for PCA-Transformed Features ---")
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

print("--- Generating Dendrogram for Hierarchical Clustering (Basic Features) ---")

# --- Step 1: Compute the linkage matrix ---
# The linkage matrix 'Z' encodes the hierarchical clustering structure.
# 'ward' linkage minimizes the variance within each cluster when merging.
# It's particularly effective for Euclidean distances.
Z = linkage(clustering_features_scaled, method='ward')

# --- Step 2: Plot the dendrogram ---
# A dendrogram is a tree-like diagram that records the sequences of merges or splits.
# It helps in visualizing the hierarchical structure of the clusters.
plt.figure(figsize=(15, 8))
dendrogram(
    Z,
    leaf_rotation=90.,  # Rotate x-axis labels to prevent overlap
    leaf_font_size=8.,  # Adjust font size for better readability
    color_threshold=None # Show all merges in different colors by default
)
plt.title('Hierarchical Clustering Dendrogram (Ward Linkage)')
plt.xlabel('Restaurant Index or Cluster Size')
plt.ylabel('Distance')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print("--- Dendrogram Generation Completed ---")
```

```python
import numpy as np
import pandas as pd
import joblib
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

class RestaurantClusteringPipeline:
    """
    Production-ready Restaurant Clustering Pipeline.

    Handles:
    - Feature Engineering
    - Scaling
    - K-Means Prediction
    - Business-Friendly Cluster Interpretation
    """

    def __init__(self, scaler: StandardScaler, kmeans_model: KMeans):
        self.scaler = scaler
        self.kmeans_model = kmeans_model

        # Business interpretation of K-Means clusters
        self.cluster_mapping = {
            0: "Ultra Budget-Friendly Specialized Restaurants",
            1: "Premium Multi-Cuisine Restaurants",
            2: "Premium High-Visibility Restaurants",
            3: "Mid-Range Mainstream Restaurants"
        }

    def predict_cluster(self,
                        cost: float,
                        cuisine_count: int,
                        collection_count: int):
        """
        Returns the numerical cluster ID.
        """

        cost_log = np.log1p(cost)
        collection_count_log = np.log1p(collection_count)

        input_data = pd.DataFrame(
            [[cost_log, cuisine_count, collection_count_log]],
            columns=[
                'Cost_log',
                'cuisine_count',
                'collection_count_log'
            ]
        )

        scaled_input = self.scaler.transform(input_data)

        cluster_id = self.kmeans_model.predict(scaled_input)[0]

        return int(cluster_id)

    def predict_cluster_name(self,
                             cost: float,
                             cuisine_count: int,
                             collection_count: int):
        """
        Returns the business-friendly cluster name.
        """

        cluster_id = self.predict_cluster(
            cost,
            cuisine_count,
            collection_count
        )

        return self.cluster_mapping[cluster_id]

    def predict_complete(self,
                         cost: float,
                         cuisine_count: int,
                         collection_count: int):
        """
        Returns both cluster ID and business segment.
        """

        cluster_id = self.predict_cluster(
            cost,
            cuisine_count,
            collection_count
        )

        cluster_name = self.cluster_mapping[cluster_id]

        return {
            "Cluster_ID": cluster_id,
            "Business_Segment": cluster_name
        }


# Create pipeline
restaurant_pipeline = RestaurantClusteringPipeline(
    scaler,
    kmeans_model
)

print("Pipeline Created Successfully")

# Save pipeline
joblib.dump(
    restaurant_pipeline,
    "RestaurantClusteringPipeline.joblib"
)

print("Pipeline Saved Successfully")
```
